```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1,data2])
```

```python
# We change the value of whis to make sure we cover all outliers
sns.boxplot(data=[data1,data2], whis=np.inf)
```

```python
# We can make it vertical using the orient parameter
sns.boxplot(data1, orient='V')
```

```python
# We can also create violin plots

data1 = stats.norm(0,5).rvs(100)

# Two gamma distributions concatenated together

data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1 * stats.gamma(5).rvs(50)])

# Box plot both data1 and data2

# These two distributions look very similar even thought they 
# are not the same distribution

sns.boxplot(data=[data1,data2], whis=np.inf)

# violin plot are like kde plots and boxplots
```

```python
sns.violinplot(data=[data1,data2])
```

```python
# We can also change the bandwidth of the kernel

sns.violinplot(data2, bw=0.01)
```

```python
# We can include individual plots 

sns.violinplot(data1, inner='stick')
```
