```python
def max_to_min(arr):
    return arr.max() - arr.min()
```

```python
# Group by object by quality
wino = dframe_wine.groupby('quality')
```

```python
wino.describe()
```

```python
# We pass a general aggregation method using .agg

# In this example this goes through wino and takes
# the max of each column value and subtracts the minimum
wino.agg(max_to_min)
```

```python
wino.agg('mean')
```
