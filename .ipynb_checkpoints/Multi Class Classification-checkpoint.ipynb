```python
# Data Imports
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

# Plot imports
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')

%matplotlib inline
```

```python
from sklearn import linear_model

from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X = iris.data

Y = iris.target
```

```python
print iris.DESCR
```

```python
iris_data = DataFrame(X, columns=['Sepal Length',
                                 'Sepal Width',
                                 'Petal Length',
                                 'Petal Width'])
iris_target=DataFrame(Y, columns=['Species'])
```

```python
iris_target.head()
```

```python
def flower(num):
    if num == 0:
        return 'Setosa'
    elif num == 1:
        return 'Versicolour'
    else:
        return 'Virginica'
```

```python
iris_target['Species'] = iris_target['Species'].apply(flower)
```

```python
iris_target.head()
```

```python
iris = pd.concat([iris_data, iris_target],axis=1)
iris.head()
```

```python
sns.pairplot(iris,hue='Species',size=2)
```

```python
sns.factorplot('Petal Length', data=iris, hue='Species',
               kind='count',size=10)
```

```python
sns.factorplot('Petal Width', data=iris, hue='Species',
               kind='count',size=10)
```

```python
sns.factorplot('Sepal Length', data=iris, hue='Species',
               kind='count',size=10)
```

```python
sns.factorplot('Sepal Width', data=iris, hue='Species',
               kind='count',size=10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
```

```python
logreg = LogisticRegression()

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,
                                                   test_size=0.3,
                                                   random_state=3)
```

```python
logreg.fit(X_train,Y_train)
```

```python
from sklearn import metrics
```

```python
Y_pred = logreg.predict(X_test)
```

```python
print metrics.accuracy_score(Y_test, Y_pred)
```

```python
# We will now use K nearest neighbors
from sklearn.neighbors import KNeighborsClassifier
```

```python
Knn = KNeighborsClassifier(n_neighbors=6)
```

```python
Knn.fit(X_train,Y_train)
```

```python
Y_pred = Knn.predict(X_test)
```

```python
print metrics.accuracy_score(Y_test, Y_pred)
```

```python
Knn = KNeighborsClassifier(n_neighbors=1)
```

```python
Knn.fit(X_train,Y_train)
```

```python
Y_pred = Knn.predict(X_test)
```

```python
print metrics.accuracy_score(Y_test, Y_pred)
```

```python
# We will find the optimal k value by cycling

k_range = range(1,21)

accuracy = []
```

```python
for k in k_range:
    Knn = KNeighborsClassifier(n_neighbors=k)
    Knn.fit(X_train,Y_train)
    Y_pred = Knn.predict(X_test)
    accuracy.append(metrics.accuracy_score(Y_test, Y_pred))
```

```python
plt.plot(k_range,accuracy)
plt.xlabel('K value')
```
