```python
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
from scipy import stats
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
```

```python
dataset2 = randn(80)

plt.hist(dataset2, color='indianred')
```

```python
# We can plot both histograms on the same plot
plt.hist(dataset1,normed=True, color='indianred',
        alpha=0.5, bins=20)
plt.hist(dataset2, normed=True, alpha=0.5, bins=20)
```

```python
# Now we do it with seaborn

data1 = randn(1000)
data2 = randn(1000)
```

```python
# Can represent as a joint dist using joint plots

sns.jointplot(data1,data2)
```

```python
sns.jointplot(data1,data2,kind='hex')
```
