```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
dataset = randn(100)
```

```python
# Allows us to combine plots. By default it uses a kde plot
# and a histogram.

sns.distplot(dataset, bins=25)
```

```python
# We can also choose differrent plots

sns.distplot(dataset, bins=25, rug=True, hist=False)
```

```python
# If we want to control specific plots within the combine plots
# we use a dictionaries

sns.distplot(dataset, bins=25, 
            kde_kws={'color':'indianred', 'label':'KDE PLOT'},
            hist_kws={'color':'blue', 'label':'HIST'})
```

```python
# We can use pandas dataobjects

from pandas import Series

ser1 = Series(dataset, name='My)Data')

ser1
```

```python
sns.distplot(ser1, bins=25)
```
