```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
lapd_df = pd.read_csv("/Users/earleaguilar/DataSets/LAPD_Arrests_2016.csv")
```

```python
lapd_df.head()
```

```python
lapd_df['Male'] = 0
lapd_df['Female'] = 0
```

```python
lapd_df.loc[lapd_df["Sex Code"] == 'M', 'Male'] = 1
lapd_df.loc[lapd_df["Sex Code"] == 'F', 'Female'] = 1
```

```python
lapd_df.head()
```

```python
crime_per_gender = lapd_df.groupby(["GRP_DESC"]).sum()[["Male", "Female"]]
```

```python
crime_per_gender.head()
```

```python
categ1 = pd.Series(crime_per_gender['Male'].index.values)
categ1 = pd.DataFrame(categ1)

counts1 = pd.Series(crime_per_gender['Male'].values)
counts1 = pd.DataFrame(counts1)

male_crime = pd.concat((categ1, counts1), axis=1)
male_crime.columns = ['Crimes', 'Count']

sns.barplot('Count',y='Crimes', data=male_crime, color='indianred')
plt.xticks(rotation=90)
```

```python
categ2 = pd.Series(crime_per_gender['Female'].index.values)
categ2 = pd.DataFrame(categ2)

counts2 = pd.Series(crime_per_gender['Female'].values)
counts2 = pd.DataFrame(counts2)

female_crime = pd.concat((categ2, counts2), axis=1)
female_crime.columns = ['Crimes', 'Count']

sns.barplot('Count',y='Crimes', data=female_crime, color='blue')
plt.xticks(rotation=90)
```
