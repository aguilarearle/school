```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip', tips)
```

```python
sns.lmplot('total_bill', 'tip', tips, 
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
# We can increase the order for a tighter fit using the 
# parameter order=

sns.lmplot('total_bill', 'tip', tips, order=4, 
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
# We dont have to fit a regression at all

sns.lmplot('total_bill','tip',tips, fit_reg=False)
```

```python
# Lm plot also works on discrete variables

tips.head()
```

```python
tips['tip_pct'] = 100 * (tips['tip']/tips['total_bill'])
tips.head()
```

```python
sns.lmplot('size','tip_pct', tips)
```

```python
# We can add a jitter for visibility

url = 'http://en.wikipedia.org/wiki/Jitter'

sns.lmplot('size', 'tip_pct', tips, x_jitter=1)
```

```python
# We can estimate the tendency of each bin

# in this case it is the size of the party

sns.lmplot('size','tip_pct', tips, x_estimator=np.mean)
```

```python
# We can use the hue facet to identify subsets on a column

sns.lmplot('total_bill', 'tip_pct', tips, hue='sex',markers=['x','o'])
```

```python
# What if we wanted to know if the day made a difference 
# in how much the tip is

sns.lmplot('total_bill','tip_pct',tips, hue='day')
```

```python
# Seaborn support local regression, model fitting

url = 'http://en.wikipedia.org/wiki/local_regression'

sns.lmplot('total_bill','tip_pct', tips, lowess=True, 
          line_kws={'color':'black'})
```

```python
# Lower level function that lmplot is built on

sns.regplot('total_bill', 'tip_pct', tips)
```

```python
fig, (axis1, axis2) = plt.subplots(1, 2,sharey=True )

sns.regplot('total_bill', 'tip_pct', tips, ax=axis1)

sns.violinplot(tips['size'],tips['tip_pct'], inner='quartile',
               cut=3, pallete='Reds_r',
               ax=axis2)
```
